```python
#Imports
import sys
sys.path.append ('../../python')
import NGC5533_functions as nf
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Check Values
print("Velocity at r=10[kpc]:",nf.bh_v(10))
```

```python
#Plot Orbital Velocity for Black Hole 
fig = plt.figure(figsize=(9.0,8.0))                #size of the plot
ax = plt.axes()
ax.grid(True)
ax.yaxis.grid(True,which='minor',linestyle='--')

r = np.linspace(0, 10, 500)

plt.plot(r, nf.bh_v(r,load=True), linestyle='solid', label='BH Velocity')          

plt.axis([0,10,0,700]);                            #xmin, xmax, ymin, ymax
                                                   #or use: plt.xlim(10, 0) plt.ylim(1.2, -1.2);
                                                   #or to tighten the bounds: plt.axis('tight')
                                                   #or use: plt.axis([0, 15000, 0.5e6, 2.5e6])

fig.suptitle('Black Hole Velocity', fontsize=18)
ax.set(title='NGC 5533')
ax.set_xlim(0,10)
ax.set_xlabel('r (kpc)', fontsize=14)
ax.set_ylabel('v (km/s)', fontsize=14)
ax.xaxis.set_tick_params(labelsize=14)
ax.yaxis.set_tick_params(labelsize=14);

plt.savefig('Outputs/BHVelocity.png')
```
