```python
#Imports
import sys
sys.path.append('../../python/')
import NGC5533_functions as nf
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Check Values
print("r-hat at r=z=1:",nf.h_rhat(1,1))
print("rho at r=1:",nf.h_rho(1,nf.h_rho00,nf.h_rc))
print("Halo Velocity at ",nf.h_vcasertano(0.001,0))
```

```python
#Variables
r = np.linspace(0, 25, 500)
```

```python
#plot orbital velocity for dark matter halo

fig = plt.figure(figsize=(9.0,8.0))                #size of the plot
ax = plt.axes()
ax.grid(True)
ax.yaxis.grid(True,which='minor',linestyle='--')

plt.plot(r, nf.h_vcasertano(r,0), linestyle='solid', label='DM Velocity')          

fig.suptitle('Dark Matter Halo - Casertano', fontsize=18)
ax.set(title='NGC 5533')              #labeling axes
#ax.set_xlim(0,25)
ax.set_xlabel('r (kpc)', fontsize=14)
ax.set_ylabel('v (km/s)', fontsize=14)
ax.xaxis.set_tick_params(labelsize=14)
ax.yaxis.set_tick_params(labelsize=14)

plt.axis([0,10,0,200])

plt.savefig('Outputs/DMVelocity.png')
```
