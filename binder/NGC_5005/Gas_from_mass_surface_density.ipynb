```python
# Import
import numpy as np
import matplotlib.pyplot as plt
```

## Trial #1

```python
# from https://academic.oup.com/mnras/article/449/4/3981/1195237#92052944 (fig.8)
# DataThief

# Radius in arcsec
gasmsd_arcsec = np.loadtxt("../data/gasmsd_arcsec_log.txt")
rarcsec, gasmsd_arcsec_log = gasmsd_arcsec.T

# Radius in kpc
gasmsd_kpc = np.loadtxt("../data/gasmsd_kpc_log.txt")
rkpc, gasmsd_kpc_log = gasmsd_kpc.T

# Change log scale to linear scale
gasmsd_arcsec_linear = np.power(10, gasmsd_arcsec_log)
gasmsd_kpc_linear = np.power(10, gasmsd_kpc_log)

# Save it to text file
# Radius in arcsec
np.savetxt("../data/gasmsd_arcsec_linear.txt", 
           np.transpose([rarcsec, gasmsd_arcsec_linear]), 
           delimiter='\t', 
           fmt='%10.5f',
           header='Datathief \n Gas mass density from https://academic.oup.com/mnras/article/449/4/3981/1195237#92052944 (fig.8) \n log scale converted to linear in python \n units: x-axis: arcsec, y-axis: mass surface density (Msun pc^-2)')

# Radius in kpc
np.savetxt("../data/gasmsd_kpc_linear.txt", 
           np.transpose([rkpc, gasmsd_kpc_linear]), 
           delimiter='\t', 
           fmt='%10.5f',
           header='Datathief \n Gas mass density from https://academic.oup.com/mnras/article/449/4/3981/1195237#92052944 (fig.8) \n log scale converted to linear in python \n units: x-axis: kpc, y-axis: mass surface density (Msun pc^-2)')
```

```python
print(gasmsd_kpc_linear)
```

```python
# Plot Gas mass surface density in kpc
plt.figure(figsize=(9,8))
plt.scatter(rkpc,gasmsd_kpc_linear)
plt.title("Gas mass surface density - Trial 1")
plt.xlabel("Radius (kpc)")
plt.ylabel("Gas mass surface density ($M_\odot$ $pc^{-2}$)")
plt.show()
```

```python
# Plot Gas mass surface density in arcsec
plt.figure(figsize=(9,8))
plt.scatter(rarcsec,gasmsd_arcsec_linear)
plt.title("Gas mass surface density - Trial 1")
plt.xlabel("Radius (arcsec)")
plt.ylabel("Gas mass surface density ($M_\odot$ $pc^{-2}$)")
plt.show()
```
